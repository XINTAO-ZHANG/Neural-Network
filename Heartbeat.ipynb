```python
import numpy
import pandas
from keras.models import Sequential
from keras.layers import Dense,Dropout, Activation
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
```

```python
# load dataset
dataframe = pandas.read_csv("heartbeats.csv", header=None)
dataset = dataframe.values
X = dataset[:,0:262].astype(int)
Y = dataset[:,262]
print (len(X))
```

```python
# Onehot encode the label
onehot_encoder = OneHotEncoder(sparse=False)
Y = Y.reshape(len(Y), 1)
Y = onehot_encoder.fit_transform(Y)
```

```python
# pick out test set, consisting 10% of the total data
idx = numpy.random.randint(len(X), size=len(X)//10)
X_test = X[idx]
Y_test = Y[idx]
# The 90% remaining will be the training set
X_train = numpy.delete(X, idx, axis=0)
Y_train = numpy.delete(Y, idx, axis=0)
```

```python
# define baseline model
def baseline_model():
    # create model
    model = Sequential()
    model.add(Dense(8, input_dim=262, activation='relu'))
    model.add(Dense(16, input_dim=8, activation='relu'))
    model.add(Dense(32, input_dim=8, activation='relu'))
    model.add(Dense(7, activation='softmax'))
    # Compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
# use estimator to pick out the best model
estimator = KerasClassifier(build_fn=baseline_model, epochs=50, batch_size=256, verbose=1)
```

```python
# split the training dataset into training set and validation set
kfold = KFold(n_splits=10, shuffle=True, random_state=seed)
```

# Use keras to pick out the best model

### 16 units, 50 epochs
92.43%

```python
results = cross_val_score(estimator, X, Y, cv=kfold)
print("Baseline: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

### 32 units, 50 epochs

88.28%

```python
results = cross_val_score(estimator, X, Y, cv=kfold)
print("Baseline: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

### 8 units, 16 units, 50 epcohs
96.46%

```python
results = cross_val_score(estimator, X, Y, cv=kfold)
print("Baseline: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

### 8 units, 16 units, 32 units, 50 epochs
<p style="color:red">97.03% - best model</p>

```python
results = cross_val_score(estimator, X, Y, cv=kfold)
print("Baseline: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

# We use the best parameters to train a model

```python
model = baseline_model()
```

```python
model.fit(X_train, Y_train, epochs=50, batch_size=256)
```

## Evaluate the model

```python
# Evaluate the model
score = model.evaluate(X_test, Y_test, batch_size=256)
```

```python
print ("Loss:%.2f Accuracy:%.2f%%" %(score[0],score[1]*100))
```
